Упражнение 1

```python
from matplotlib.pyplot import *
from math import *
```

```python
def f(x):
    # простая функция f(x) = -0.5*x + 1.5
    return -0.5 * x + 1.5

# начальное приближение
x0 = 2.5

# списки для точек ломаной
x_points = [x0]  # координаты x
y_points = [0]   # координаты y

# строим ломаную из 4 итераций
x = x0
for i in range(4):
    x_points.append(x)    # вертикальная линия
    y_points.append(f(x)) 
    x_points.append(f(x)) # горизонтальная линия
    y_points.append(f(x))
    x = f(x)              # новое значение для следующей итерации

# рисуем ломаную
figure(figsize=(10, 8))

# строим график функции y = f(x) по точкам
x_vals = [i/10 for i in range(-5, 31)]  # от -0.5 до 3 с шагом 0.1
y_vals = [f(x) for x in x_vals]
plot(x_vals, y_vals, 'b-', linewidth=2, label='y = f(x)')

# строим прямую y = x
plot(x_vals, x_vals, 'r--', linewidth=2, label='y = x')

# рисуем ломаную итерационного процесса
plot(x_points, y_points, 'g-', linewidth=1.5, alpha=0.7)

# отмечаем точки на ломаной
# красные точки - на прямой y=x
scatter(x_points[::2], x_points[::2], color='red', s=30, zorder=5)
# синие точки - на графике функции
scatter(x_points[1::2], y_points[1::2], color='blue', s=30, zorder=5)

# выделяем отрезок [a; b]
axvspan(0.2, 2.2, alpha=0.1, color='gray', label='отрезок [a; b]')

# настройки графика
grid(True, alpha=0.3)
axhline(y=0, color='k', linewidth=0.5)  # ось x
axvline(x=0, color='k', linewidth=0.5)  # ось y
xlabel('x', fontsize=12)
ylabel('y', fontsize=12)
title('Сходящийся итерационный процесс', fontsize=14)
legend(fontsize=12)
axis('equal')  # одинаковый масштаб по осям

tight_layout()
show()
```

Упражнение 2

```python
def f(x):
    # итерационная функция f(x) = 1/sin(x)
    return 1 / sin(x)

x = 1.0 # начальное приближение
n = 0 # счётчик
# итерационный процесс
while True:
    n += 1
    x_new = f(x)  # новое приближение
    print(f"Итерация {n}: x = {x_new:.8f}")
    
    # проверка точности для окончания цикла
    if abs(x_new - x) < 0.00001:
        print(f"Итерация {n} - где мы достигли нужной точности")
        print(f"Значение корня: {x_new:.8f}")
        break
    
    x = x_new  # подготовка к следующей итерации
```

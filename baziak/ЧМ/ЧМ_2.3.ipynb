```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -0.5*x + 1.5
xi = 1.0

x0 = 2.5

x_points = [x0]
y_points = [0]

x = x0
for i in range(4):
    x_points.append(x)
    y_points.append(f(x))
    x_points.append(f(x))
    y_points.append(f(x))
    x = f(x)

plt.figure(figsize=(10, 8))

X = np.linspace(-0.5, 3, 400)
Y_f = f(X)
Y_x = X

plt.plot(X, Y_f, 'b-', linewidth=2, label='y = f(x)')
plt.plot(X, Y_x, 'r--', linewidth=2, label='y = x')

plt.plot(x_points, y_points, 'g-', linewidth=1.5, alpha=0.7)

plt.scatter(x_points[::2], x_points[::2], color='red', s=30, zorder=5)
plt.scatter(x_points[1::2], y_points[1::2], color='blue', s=30, zorder=5)

plt.axvspan(0.2, 2.2, alpha=0.1, color='gray', label='Отрезок [a; b]')

plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title(f'Сходящийся итерационный процесс', fontsize=14)
plt.legend(fontsize=12)
plt.axis('equal')


plt.tight_layout()
plt.show()

print(f"Производная функции f'(x) = -0.5 на всем отрезке")
print(f"Условие -1 < f'(x) < 0 выполняется: -1 < -0.5 < 0")
```

```python
def form(x):
    return 1 / np.sin(x)

x = 1.0

for n in range(1, 100):
    x_new = form(x)
    print(f"Итерация {n}: x = {x_new:.8f}")
    if abs(x_new - x) < 0.00001:
        print(f"Нужная точность на итерации {n}")
        print(f"Значение корня: {x_new:.8f}")
        print(f"Проверка: {x_new * np.sin(x_new) - 1:.8f}")
        break
    x = x_new
```

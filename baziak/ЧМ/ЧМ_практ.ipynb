Задание 1 - граф

```python
from math import *  # импортируем все математические функции
from matplotlib.pyplot import *

def f(x):
    # Исходное уравнение lg(x+5) - cos(x)
    return log10(x + 5) - cos(x)

def f1(x):
    # Функция lg(x+5)
    return log10(x + 5)

def f2(x):
    # Функция cos(x)
    return cos(x)

# список значений x от -4.9 до 4.9
x_vals = []
step = (4.9 - (-4.9)) / 1000  # шаг между точками
x = -4.9
for i in range(1001):  # 1001 точка для 1000 интервалов
    x_vals.append(x)
    x += step

# значения функций для каждого x
y1_vals = [f1(x) for x in x_vals]  # lg(x+5)
y2_vals = [f2(x) for x in x_vals]  # cos(x)

# график
figure(figsize=(10, 6))
plot(x_vals, y1_vals, label='lg(x+5)', linewidth=2)
plot(x_vals, y2_vals, label='cos(x)', linewidth=2)
axhline(0, color='black', linewidth=0.5)  # горизонтальная линия y = 0
axvline(-5, color='red', linestyle='--', label='x > -5')  # вертикальная линия x = -5
grid(True, alpha=0.3)
xlabel('x')
ylabel('y')
title('lg(x+5) = cos(x)')
xlim(-6, 6)
ylim(-2, 2)

# интервал где функции пересекаются
roots_intervals = []
for i in range(len(x_vals) - 1):
    diff1 = y1_vals[i] - y2_vals[i] # разность в текущей точке
    diff2 = y1_vals[i+1] - y2_vals[i+1] # разность в следующей точке
    
    # eсли знаки разности разные - eсть корень
    if diff1 * diff2 < 0:
        x1, x2 = x_vals[i], x_vals[i+1]
        roots_intervals.append((x1, x2))
        print(f"Корень в интервале: {x1:.3f}, {x2:.3f}")

legend()
show()
```

Задание 2 - метод половинного деления

```python
def f(x):
    # исходная функция f(x) = lg(x+5) - cos(x)
    return log10(x + 5) - cos(x)

# задаем границы интервала a, b и точность eps
a, b, eps = 0.6, 0.8, 0.001
n = 0  # счетчик итераций

# метод половинного деления
# продолжаем делить интервал пока его длина больше точности
while b - a > eps:
    # находим середину интервала
    c = (a + b) / 2
    
    # проверяем знаки функции на концах [a, c]
    # если f(a)*f(c) меньше 0 - значит корень между a и c
    if f(a) * f(c) <= 0:
        b = c  # сдвигаем право к середине
    else:
        a = c  # иначе сдвигаем лево к середине
    
    n += 1  # счетчик итераций

# выводим результаты
print(f"Номер итерации: {n}")
print(f"Корень: x = {c:.6f}")
print(f"f(x) = {f(c):.6f}")
```

Задание 2 - метод хорд

```python
def f(x):
    # функция lg(x+5) - cos(x)
    return log10(x + 5) - cos(x)

def F(x):
    # Функция F(x)
    return log10(x + 5) - cos(x)

def dF(x):
    # производная функции f(x)
    return 1/((x + 5) * log(10)) + sin(x)

# интервал a, b - где есть корень и точность
a, b, eps = 0.6, 0.8, 0.001

# придаём значение функций переменным
fa = F(a)
fb = F(b)

# проверка знаков
print(f"F({a}) = {fa:.6f}")
print(f"F({b}) = {fb:.6f}")

# определяем неподвижную точку c 
# знаки разные - можно взять любой конец
if fa * F((a + b) / 2) > 0:
    c = a  # неподвижная точка
    x0 = b  # начальное приближение
    print(f"Неподвижная точка c = {c}")
else:
    c = b
    x0 = a
    print(f"Неподвижная точка c = {c}")


n = 1
x_prev = x0  # предыдущее приближение
x_curr = c  # текущее приближение
f_curr = f(x_curr)  # значение в текущей точке

while True:
    f_prev = f(x_prev)  # значение в предыдущей точке
    
    # вычисляем следующее значение по формуле метода хорд
    x_next = x_prev - f_prev * (x_prev - x_curr) / (f_prev - f_curr)
    
    print(f"Итерация {n}: x = {x_next:.6f}, f(x) = {f(x_next):.8f}")
    
    # проверка достижения точности, если достигли, цикл выключается
    if abs(x_next - x_prev) < eps:
        break
    
    # Подготовка к следующей итерации в случае неостановки
    x_curr = x_prev
    f_curr = f_prev
    x_prev = x_next
    n += 1 # счёт итераций

# вывод
print(f"Число итераций: {n}")
print(f"Найденный корень: x = {x_next:.6f}")
print(f"Значение функции f(x) = lg(x+5) - cos(x) = {f(x_next):.8f}")
```

Задание 3 - метод касательных

```python
def f(x):
    # функция lg(x+5) - cos(x)
    return log10(x + 5) - cos(x)

def dF(x):
    # производная функции f'(x)
    return 1/((x + 5) * log(10)) + sin(x)

# задаем границы интервала a, b и точность
a, b, eps = 0.6, 0.8, 0.001
n = 1  # счетчик итераций

# выбираем начальное приближение x0 для метода ньютона
# берем тот конец, где знак функции совпадает со знаком второй производной

if f(a) * dF(a) > 0: # если условие выполняется на левом конце
    x0 = a 
elif f(b) * dF(b) > 0: # если на правом конце
    x0 = b  
else:  # если нигде не выполняется берем середину
    x0 = (a + b) / 2

x = x0 # текущее приближение

# находим минимальное значение функции на концах интервала
# это нужно для оценки погрешности
m = min(abs(f(a)), abs(f(b)))

print(f"Итерация {n}: x = {x:.6f}, f(x) = {f(x):.6f}")

# метод ньютона (касательных)
while True:
    # решаем по формуле метода ньютона
    x_next = x - f(x) / dF(x)
    fN = f(x_next)  # значение функции в новой точке
    
    print(f"Номер итерации {n+1}: x = {x_next:.6f}, f(x) = {fN:.6f}")
    
    x = x_next # обновляем текущее приближение
    n += 1 # увеличиваем счетчик
    
    # мы достигли нужной точности и заканчиваем while
    if abs(fN) / m < eps:
        break

print(f"Номер итерации: {n}")
print(f"Корень: x = {x:.6f}")
print(f"f(x) = lg(x+5) - cos(x) = {f(x):.6f}")
```
